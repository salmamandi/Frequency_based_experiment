```python
import pandas as pd
import numpy as np

for i in range(1,23,1): 
        if(i==9 or i==19 or i==21):
            continue
        else:    
            file1="SMOTE_folder/Time_dataset_new_ICASSP/"+"U"+str(i)+"_LIHF_itd_file.csv"
            print(file1)
            #file2="SMOTE_folder/All_Amp/"+"user_"+str(i)+"_v4.csv"
            file2="SMOTE_folder/All_amp/"+"U"+str(i)+"_LIHF_itd_file.csv"
            print(file2)
            data1=pd.read_csv(file1)
            data2=pd.read_csv(file2,header=None)
    
            array1=data1.values
            array2=data2.values
            col_no=array2.shape[1]-1
        
            write_data1=array2[:,0:col_no]
            write_data2=array1[:,0:5]
            print(write_data1.shape)
    
            print(write_data2.shape)
        
            raw_data=np.append(write_data1,write_data2,axis=1)
            print(type(raw_data))
            #np.savetxt("SMOTE_folder/TD+FD/"+"U"+str(i)+"_LIHF_itd_file.csv", raw_data, delimiter=",")
    
            #data_file=pd.DataFrame.from_records(raw_data,columns=['Amp1','Amp2','MSI','first','second','third','label'])
            data_file=pd.DataFrame.from_records(raw_data)
            with open("SMOTE_folder/Allamp_timeDomain/"+"U"+str(i)+"_LIHF_itd_file.csv",'w') as result:
                data_file.to_csv(result,index=False)
```
