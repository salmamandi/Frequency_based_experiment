```python
import pandas as pd
import os

for root, dirs,files in os.walk("raw_Data/"):
    for directory in dirs:
        dir_name="raw_Data/"+directory
        print(dir_name)
        for filename in os.listdir(dir_name):
            print(filename)
            text_file=dir_name+"/"+filename
            read_file=pd.read_csv(text_file)
            read_file.to_csv("CSV_raw_data/"+directory+"/"+filename+".csv",index=None)
```
